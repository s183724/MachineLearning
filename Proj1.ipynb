Creating a new column, called continent. Here we will use the library pycountry to check upon the alpha codes, and thereby inserting the correct continent.

```python
continents = {
    'NA': 'North America',
    'SA': 'South America', 
    'AS': 'Asia',
    'OC': 'Australia',
    'AF': 'Africa',
    'EU': 'Europe',
}
dfs['Continent']= [continents[country_alpha2_to_continent_code(country_name_to_country_alpha2(country))] for country in dfs.Country]
dfs
```

```python
import pycountry
import plotly.express as px

list_countries = df.index.values.tolist()
d_country_code = {} 
for country in list_countries:
    try:
        country_data = pycountry.countries.search_fuzzy(country)
        country_code = country_data[0].alpha_3
        d_country_code.update({country: country_code})
    except:
        d_country_code.update({country: ' '})

for k, v in d_country_code.items():
    df.loc[(df.index.values == k), 'iso_alpha'] = v

fig = px.choropleth(data_frame = df,
                    locations= "iso_alpha",
                    color= "APC",
                    hover_name= df.index.values,
                    color_continuous_scale= 'YlOrRd')

fig.show()
```
